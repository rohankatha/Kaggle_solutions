```python
# Import libraries
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import numpy as np

# Generate some dummy regression data
# 1000 samples, 5 input features
X = np.random.random((1000, 5))
y = np.random.random((1000, 1))  # 1 output (regression)

# Define the Model
input_layer = Input(shape=(5,))   # 5 inputs
hidden_1 = Dense(20, activation='relu')(input_layer)   # first dense with 20 neurons
hidden_2 = Dense(50, activation='relu')(hidden_1)      # then 50
hidden_3 = Dense(10, activation='relu')(hidden_2)      # then 10
output = Dense(1, activation='linear')(hidden_3)       # output with 1 neuron (linear for regression)

model = Model(input_layer, output)

# Compile the Model
model.compile(optimizer='adam', loss='mse', metrics=['mse'])

# Train the Model
history = model.fit(X, y, epochs=5, batch_size=32, validation_split=0.2)

# Evaluate the Model
loss, mse = model.evaluate(X, y)
print(f"Model evaluation - Loss (MSE): {loss}")

# Summary
model.summary()
```
